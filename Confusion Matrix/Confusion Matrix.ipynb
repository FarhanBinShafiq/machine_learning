```python
# Import necessary libraries
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example 1: Basic Confusion Matrix
# Let's create some sample data
actual_labels = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0]    # True labels
predicted_labels = [1, 0, 1, 0, 1, 1, 0, 1, 1, 0]  # Predicted by model
```

```python
# Create confusion matrix
cm = confusion_matrix(actual_labels, predicted_labels)
```

```python
print("Basic Confusion Matrix:")
print(cm)
print("\nComponents:")
print("True Negatives (TN):", cm[0][0])  # Top-left
print("False Positives (FP):", cm[0][1]) # Top-right
print("False Negatives (FN):", cm[1][0]) # Bottom-left
print("True Positives (TP):", cm[1][1])  # Bottom-right
```

```python
# Example 2: Visualizing Confusion Matrix with Heatmap
# Creating a more realistic example with multiple classes
y_true = ['cat', 'dog', 'bird', 'cat', 'dog', 'bird', 'cat', 'dog']
y_pred = ['cat', 'dog', 'cat', 'cat', 'dog', 'bird', 'dog', 'cat']
```

```python
# Generate confusion matrix
cm_multi = confusion_matrix(y_true, y_pred)
class_names = ['cat', 'bird', 'dog']  # alphabetical order
```

```python
# Plot using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm_multi, annot=True, fmt='d', cmap='Blues',
           xticklabels=class_names,
           yticklabels=class_names)
plt.title('Confusion Matrix - Animal Classification')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
# Example 3: Calculating metrics from confusion matrix
def calculate_metrics(cm):
    TN = cm[0][0]
    FP = cm[0][1]
    FN = cm[1][0]
    TP = cm[1][1]
    
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)
    
    return accuracy, precision, recall, f1_score
```

```python
# Calculate metrics for binary example
acc, prec, rec, f1 = calculate_metrics(cm)
print("\nMetrics for Binary Classification:")
print(f"Accuracy: {acc:.2f}")
print(f"Precision: {prec:.2f}")
print(f"Recall: {rec:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
# Import necessary libraries
from sklearn.metrics import confusion_matrix  # Gets the confusion matrix tool from scikit-learn
import numpy as np                           # Math library (like a calculator)
import matplotlib.pyplot as plt              # For making basic plots
import seaborn as sns                        # For making prettier plots

# Example 1: Basic Confusion Matrix
actual_labels = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0]    # Real answers (1 = yes, 0 = no)
predicted_labels = [1, 0, 1, 0, 1, 1, 0, 1, 1, 0]  # What our model guessed

cm = confusion_matrix(actual_labels, predicted_labels)  # Makes a table comparing real vs. guessed

print("Basic Confusion Matrix:")    # Shows text on screen
print(cm)                          # Shows the table
print("\nComponents:")             # \n makes a new line
print("True Negatives (TN):", cm[0][0])   # Top-left: correct "no" guesses
print("False Positives (FP):", cm[0][1])  # Top-right: wrong "yes" guesses
print("False Negatives (FN):", cm[1][0])  # Bottom-left: wrong "no" guesses
print("True Positives (TP):", cm[1][1])   # Bottom-right: correct "yes" guesses

# Example 2: Visualizing Confusion Matrix
y_true = ['cat', 'dog', 'bird', 'cat', 'dog', 'bird', 'cat', 'dog']  # Real animal labels
y_pred = ['cat', 'dog', 'cat', 'cat', 'dog', 'bird', 'dog', 'cat']   # Guessed animal labels

cm_multi = confusion_matrix(y_true, y_pred)  # Makes table for animal guesses
class_names = ['cat', 'bird', 'dog']         # List of animal names

plt.figure(figsize=(8, 6))                   # Sets up a blank graph 8 wide, 6 tall
sns.heatmap(cm_multi, annot=True, fmt='d', cmap='Blues',  # Makes a colorful table
           xticklabels=class_names,              # Labels columns with animal names
           yticklabels=class_names)              # Labels rows with animal names
plt.title('Confusion Matrix - Animal Classification')     # Adds title
plt.ylabel('True Label')                     # Labels vertical axis
plt.xlabel('Predicted Label')                # Labels horizontal axis
plt.show()                                  # Shows the graph

# Example 3: Calculating metrics
def calculate_metrics(cm):                  # Creates a function to do math
    TN = cm[0][0]                          # Gets True Negatives
    FP = cm[0][1]                          # Gets False Positives
    FN = cm[1][0]                          # Gets False Negatives
    TP = cm[1][1]                          # Gets True Positives
    
    accuracy = (TP + TN) / (TP + TN + FP + FN)  # How often we're right
    precision = TP / (TP + FP)                 # How good our "yes" guesses are
    recall = TP / (TP + FN)                    # How many "yes" we caught
    f1_score = 2 * (precision * recall) / (precision + recall)  # Mix of precision and recall
    
    return accuracy, precision, recall, f1_score  # Gives back all numbers

acc, prec, rec, f1 = calculate_metrics(cm)    # Uses function on our first table
print("\nMetrics for Binary Classification:") # Shows results
print(f"Accuracy: {acc:.2f}")                # Shows accuracy (2 decimal places)
print(f"Precision: {prec:.2f}")              # Shows precision
print(f"Recall: {rec:.2f}")                  # Shows recall
print(f"F1 Score: {f1:.2f}")                 # Shows F1 score
```
