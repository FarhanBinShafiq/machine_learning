```python
import pandas as pd
```

```python
data={'categories':['A','B','c','A','B','c','A','B','c','A','B','c','A','B','c','A','B','c','A','B','c','D']}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
one_hot_encoded=pd.get_dummies(df,columns=['categories'])
one_hot_encoded
```
