```python
df=pd.read_csv('car data.csv')
df
```

```python
#null value check
df.isnull().sum()
```

### We have fill up the NaN value by that group's meadian or mean Value..

```python
df.car_age
```

```python
#meadin
car_null=df.car_age.median()
```

```python
car_null
```

```python
## NaN value fillup
df.car_age=df.car_age.fillna(car_null)
df.car_age
```

```python
###for experence and risk column

#experience
experience_null=df.experience.median()
experience_null
```

```python
df.experience=df.experience.fillna(experience_null)
df.experience
```

```python
df
```

```python
#Linear Regresssion
reg = LinearRegression()
```

```python
reg.fit(df[['speed','car_age','experience']],df.risk)
```

## Now predict for 160 km speed ,10 years old car,5 year driver experience
