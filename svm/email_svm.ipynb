```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt

# Step 1: Create sample data
# Features: [frequency of "free", number of exclamation marks]
X = np.array([[0, 1], [1, 0], [5, 3], [2, 2], [4, 4], [0, 0]])  # Training data
y = np.array([0, 0, 1, 1, 1, 0])  # Labels: 0 = Not Spam, 1 = Spam

# Step 2: Train the SVM model
clf = svm.SVC(kernel='linear')  # Using a linear kernel for simplicity
clf.fit(X, y)

# Step 3: Predict a new email
new_email = np.array([[3, 2]])  # New email with "free" 3 times, 2 exclamation marks
prediction = clf.predict(new_email)
print("Prediction (0 = Not Spam, 1 = Spam):", prediction[0])

# Step 4: Visualize the decision boundary (optional)
# Get the separating hyperplane
w = clf.coef_[0]  # Weights
w
b = clf.intercept_[0]  # Bias
b
x_line = np.linspace(-1, 6)  # Range for plotting
y_line = -(w[0] * x_line + b) / w[1]  # Equation of the hyperplane

# Plot data points and hyperplane
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', label='Data Points')
plt.plot(x_line, y_line, 'k-', label='Decision Boundary')
plt.scatter(new_email[0, 0], new_email[0, 1], c='green', marker='x', s=100, label='New Email')
plt.xlabel('Frequency of "free"')
plt.ylabel('Number of Exclamation Marks')
plt.legend()
plt.show()
```
